```python
import numpy as np
import matplotlib.pyplot as plt

xpoints = np.array([2,3,8,7])
ypoints = np.array([1,4,1,9])
plt.plot(xpoints,ypoints, marker="*", color="blue", linestyle = "dotted")

# Labelling X-axis
plt.xlabel("X-axis")
# labelling Y-axis
plt.ylabel("Y-axis")
# Labelling Whole Chart
plt.title("Graph")

# Command to plot graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import random as random

students = ["Shubham","Raj","Suketu","Nishang","Vatsal","Jay","Vinayak","Jyot"]

marks=[]
for i in range(0,len(students)) :
    marks.append(random.randint(0,101))

plt.xlabel("Students")
plt.xticks(rotation=45)
plt.ylabel("Marks")
plt.title("CLASS RECORDS")
plt.plot(students,marks,marker="o",color="black",markerfacecolor="blue",markersize=12)
plt.show()
```

```python
# Importing Matplotlib module
import matplotlib.pyplot as plt

# x-axis co-ordinates
x = [1,2,3]
# corresponding y-axis co-ordinates
y = [2,4,1]

# plotting the points
plt.plot(x,y)

# Labelling x-axis 
plt.xlabel("X-axis")
# Labelling y-axis
plt.ylabel("Y-axis")
# Labelling Graph
plt.title("Graph")

# Command for printing the output
plt.show()
```

```python
# Syncing Matplotlib with numpy library

# Importing NumPy
import numpy as np
# Importing Matplotlib
import matplotlib.pyplot as plt

# Taking the co-ordinates of x and y
x = np.arange(0,10,0.1)
y = np.sin(x)

# Manipulating Height and Width of graphs in matplotlib using figure()
# Syntax : figsize = (width, height)
plt.figure(figsize = (6,4))

# Ploting co-ordinates of x and y
plt.plot(x,y)

# Printing the output
plt.show()
```

```python
# Sub plot

# Plot 1:
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(2,2,1)
plt.plot(x,y)

# Plot 2:
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(2,2,2)
plt.plot(x,y)

#Plot 3 :
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(2,2,3)
plt.plot(x,y)

# Plot 4
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(2,2,4)
plt.plot(x,y)
# Synatx : subplot(no. of rows, no. of coloums, plot no.)
#lt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

```python
# Bar Graph plotting

# A bar graph is a chart that represents categorical data with rectangular bars with height or length proportional to the values that they represent
# The bars can be plotted vertically or horizontally.

# Syntax :
    # ax.bar(x, height, width, bottom, align)
    # plt.bar(x,y)
    
# Bar graphs are of two types :
    # 1. Vertical 
    # 2. Horizontal
# By default the type of Bar graph is Vertical
```

```python
import matplotlib.pyplot as plt
#from matplotlib import pyplot as plt

x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.bar(x,y)
plt.show()
```

```python
# Scatter Plot

# Importing matplotlib module
import matplotlib.pyplot as plt

# Co-ordinates of x-axis 
x = [5,2,9,4,7,3,8,10,1]
# Co-ordinates of y-axis
y = [10,5,8,4,2,1,6,3,9]

# Function to plot scatter graph
plt.scatter(x,y)

# Printing the ouptut
plt.show()
```

```python
# Ticks are the markers denoting data points on either axis
# Importing required libraries
import matplotlib.pyplot as plt

# Taking Co-ordinates of X and Y
x = [5,10,15,20,25,30,35,40,45,50]
y = [1,4,3,2,7,6,9,8,10,5]

plt.plot(x,y,"b")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.tick_params(axis="x", labelsize = 12, labelrotation = 20, labelcolor = "red")
plt.tick_params(axis="y", labelsize = 12, labelrotation = 20, labelcolor = "blue")

plt.show()
```

```python
# Adding Legend in the Graph

# Importing required modules
import numpy as np
import matplotlib.pyplot as plt

# Taking co-ordinates
y1 = [2,3,4.5]
y2 = [1,1.5,5]

plt.plot(y1,"blue")
plt.plot(y2,"red")

plt.legend(["blue","red"],loc ="upper right")

plt.show()
```

```python
# Histogram 
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(170,10,250)
#print(data)

plt.hist(data)
plt.show()

plt.hist(data, bins=30)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Draw plot as figure
fig, ax = plt.subplot()

x = np.linspace(0,20,1000)
ax.plot(x,np.cos(x))
ax.axis("equal")

ax.annotate('local maximum', xy = (6.8,1), xytext = (10,4), arrowprops = dict(facecolor="black", shrink=0.05))
ax.annotate('local minimum', xy = (5* np.pi,-1), xytext=(2,-6),arrowprops=dict(arrowstyle='->', connectionstyle="angle3,angleA=0,angleB=-90"))
```
