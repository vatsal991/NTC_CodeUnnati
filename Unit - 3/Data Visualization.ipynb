```python
import matplotlib.pyplot as plt
import random as random

students = ["Shubham","Raj","Suketu","Nishang","Vatsal","Jay","Vinayak","Jyot"]

marks=[]
for i in range(0,len(students)) :
    marks.append(random.randint(0,101))

plt.xlabel("Students")
plt.xticks(rotation=45)
plt.ylabel("Marks")
plt.title("CLASS RECORDS")
plt.plot(students,marks,marker="o",color="black",markerfacecolor="blue",markersize=12)
plt.show()
```

```python
# Ticks are the markers denoting data points on either axis
# Importing required libraries
import matplotlib.pyplot as plt

# Taking Co-ordinates of X and Y
x = [5,10,15,20,25,30,35,40,45,50]
y = [1,4,3,2,7,6,9,8,10,5]

plt.plot(x,y,"b")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.tick_params(axis="x", labelsize = 12, labelrotation = 20, labelcolor = "red")
plt.tick_params(axis="y", labelsize = 12, labelrotation = 20, labelcolor = "blue")

plt.show()
```

```python
# Adding Legend in the Graph

# Importing required modules
import numpy as np
import matplotlib.pyplot as plt

# Taking co-ordinates
y1 = [2,3,4.5]
y2 = [1,1.5,5]

plt.plot(y1,"blue")
plt.plot(y2,"red")

plt.legend(["blue","red"],loc ="upper right")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Draw plot as figure
fig,ax = plt.subplots()

x = np.linspace(0,20,1000)
ax.plot(x,np.cos(x))
ax.axis("equal")

ax.annotate('local maximum', xy = (6.8,1), xytext = (10,4), arrowprops = dict(facecolor="black", shrink=0.05))
ax.annotate('local minimum', xy = (5* np.pi,-1), xytext=(2,-6),arrowprops=dict(arrowstyle='->', connectionstyle="angle3,angleA=0,angleB=-90"))
```

# Adding Annotate & legend

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,10)
y=x^2
z=x^3

print(x,y,z)

plt.plot(x,y,marker="o")
plt.plot(x,z)
plt.plot(y,z)

#Adding title
plt.title("MY GRAPH")

#Adding X&Y label
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

#Annotation

plt.annotate(xy=[2,0],xytext = (1.5,3),text="First Entry",arrowprops = dict(facecolor="black", shrink=0.05),rotation="30")
plt.annotate(xy=[5,7],xytext = (5,9),text="Second Entry",arrowprops = dict(facecolor="black", shrink=0.05))
plt.annotate(xy=[8,10],xytext = (5,11),text="Third Entry",arrowprops = dict(facecolor="black", shrink=0.05))

plt.legend(["Race1", "Race2", "Race3 "],loc="lower right")
plt.show()
```

# Pie Plot

pie chart a circular statical plot that can disaply only one series of data


```python
import numpy as np
import matplotlib.pyplot as plt

#creating dataset

Branch=["Computer","Electrical","Civil","Mechenical","Automobil"]
Data=[50,34,40,29,12]
explode = (0.1,0,0,0,0)

#Create Chart

fig=plt.figure(figsize=(10,7))
plt.pie(Data,labels=Branch,explode=explode,autopct="%1.1f%%",shadow=True)
plt.show()
```

# 3D Ploting

```python
# importing  mplot3d toolkit, numpy & matplotlib

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d # 3d toolkit lib

fig = plt.figure()

#syntax for D projection
ax=plt.axes(projection = '3d')

#defining all 3 axes

z=np.linspace(0,1,100)
x=z*np.sin(25*z)
y=z*np.cos(25*z)

#3d ploting

ax.plot3D(x,y,z,'red')
ax.set_title('3D Ploting')
plt.show()
```
